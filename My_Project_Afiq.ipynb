<h4> IMPORTING LIBRARIES

```python
import pandas as pd #For data manipulation

#For data visualization
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pop = pd.read_csv('../World_Population/AsiaPopulation2020.csv')#Importing Dataset
```

<h4> CLEANING DATA

```python
pop.shape #To know our total number of rows and columns
```

```python
pop #Display the full data in table
```

```python
pop.isnull().sum() #To check if there is a null value, the output shows 3
```

```python
missing_value=["na",np.nan]#python cannot detect what is 'na'
pop=pd.read_csv('../World_Population/AsiaPopulation2020.csv', na_values=missing_value)
```

```python
pop.isnull().sum() #Actual total missing value
```

```python
pop.fillna(0.0, inplace = True) # Didnt want to delete the entire data as it will change the original value
#i just add value 0.00 to the missing value.
```

```python
pop.isnull().sum() #The output
```

<h4> Data Visualization

```python
pop_top = pop.sort_values(by='Population', ascending=False).head()
pop_top #Print out the top 5 country according to the population
```

```python
sns.barplot(data=pop_top, x='Country', y='Population')
plt.show() #Plotting the graph top 5 country according to the population
```

```python
pop_bottom = pop.sort_values(by='Population', ascending=False).tail()
pop_bottom #Print out the bottom 5 country according to the population
```

```python
sns.barplot(data=pop_bottom, x='Country', y='Population')
plt.show()#Plotting the graph for bottom 5 country according to the population
```

```python
pop_sea=pop.iloc[[6,28,7,44,13,22,25,33,36,43,49],:]
pop_sea #Group the country included in South East Asia region
```

```python
# Size of population for South East Asia country
sns.relplot(x="Country", y="Population", kind="line", data=pop_sea, height=3, aspect=30/7,)
```

```python
##Population Density Comparison Of South East Asia
sns.relplot(x="Country", y="Density", kind="line", data=pop_sea, height=3, aspect=30/7,)
```

```python
#Size of country comparison in South East Asia
sns.relplot(x="Country", y="LandArea", kind="line", data=pop_sea,height=3, aspect=30/7,)
```

```python
#Rate of Fertility in South East Asia
sns.relplot(x="Country", y="FertRate", kind="line", data=pop_sea,height=3, aspect=30/7,)
```
